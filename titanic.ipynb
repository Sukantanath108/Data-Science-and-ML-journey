```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
new_df = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
new_df.head()
```

```python
new_df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
new_df['Sex'] = le.fit_transform(new_df['Sex'])
new_df.tail()
```

```python
new_df['Age'] = new_df['Age'].fillna(new_df['Age'].mean())
new_df.head(20)
```

```python
# scaler = StandardScaler()
# new_df[['Age', 'Fare']] = scaler.fit_transform(new_df[['Age', 'Fare']])
# new_df.head()
```

```python
X = new_df.drop('Survived',axis='columns')
y = new_df['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
score = dt.score(X_test,y_test)
print("The accuracy score is :",score)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
score1 = dt.score(X_test,y_test)
print("The accuracy score is :",score1)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
score2 = dt.score(X_test,y_test)
print("The accuracy score is :",score2)
```
